```python
import numpy as np
import matplotlib.pyplot as plt

# 
sampling_rate = 44100  #  (Hz)
frequency = 1000       #  (Hz)
duration = 0.016       #  ()

# 
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

# 
pure_tone = np.sin(2 * np.pi * frequency * t)

# 
hanning_window = np.hanning(len(pure_tone))
windowed_signal = pure_tone * hanning_window

# 
plt.figure(figsize=(4, 3))

# 
plt.subplot(2, 1, 1)
plt.plot(t, pure_tone, label='Pure Tone')
plt.title(f'Pure Tone (Frequency = {frequency} Hz, Duration = {duration * 1000} ms)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
# plt.grid()
# plt.legend()

# 
plt.subplot(2, 1, 2)
plt.plot(t, windowed_signal, label='Windowed Signal')
plt.title(f'Windowed Signal (Hanning Window)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
# plt.grid()
# plt.legend()

# 
plt.tight_layout()
plt.show()
```
